```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import texttable as tt
```

```python
def read_file(filename):
    with open(filename,'r') as fm:
        lines=[f[:-1] for f in fm.readlines()]
    return lines
```

```python
mvratings=read_file('hw8_ratings.txt')
mvnames=read_file('hw8_movies.txt')
mlen=(len(mvnames))
```

```python
def split_lines(mvlist):
    output=[]
    for entry in mvlist:
        output.append(entry.split(' '))
    return output
```

```python
mvr=split_lines(mvratings)
```

```python
def movie_ratings(mvindex,mvrates):
    smvratings=[x[mvindex] for x in mvrates]
    agg_rating=Counter(smvratings)
    #print(agg_rating)
    return agg_rating['1']/(agg_rating['0']+agg_rating['1']),smvratings  
```

```python
rec_ratio=[]
for i in range(0,len(mvnames)):
    agg,smvratings=movie_ratings(i,mvr)
    rec_ratio.append(agg)
```

```python
d={x:y for x,y in zip(mvnames,rec_ratio)}
sorted_mv_rec=sorted(d, key=d.__getitem__)
```

```python
tab=tt.Texttable()
tab.header(['Movie Name','Ratio'])
for x in sorted_mv_rec:
    row=[x,d[x]]
    tab.add_row(row)
tab.set_cols_align(['c','c'])
s=tab.draw()
print(s)
```

```python
probZ=read_file('hw8_probZ_init.txt')
probR=read_file('hw8_probR_init.txt')
```

```python
def split_space(arr):
    pbZ=[]
    for val in arr:
        pbZ.append(float(val.split('   ')[1]))
    return pbZ
```

```python
pbR=np.zeros((76,4))
for i,val in enumerate(probR):
    pbR[i,0:5]=val.split('   ')[1:5]
    #print(val.split('   ')[1:5])
```

```python
total=len(mvratings)
pbZ=np.array(pbZ)
pbZ=pbZ.reshape((-1,1))
pbRzero=1-pbR
print(pbRzero)
```

```python
#likelihood of the tth student rating
def lhood_student_rating(pbZ,pbR,mvratings,pbRzero,student_index,z_index):
    in1=[]
    in0=[]
    for rating in mvratings:
        in1.append([i for i,x in enumerate(rating.split(' ')) if x=='1'])
        in0.append([i for i,x in enumerate(rating.split(' ')) if x=='0'])
    sm=0
    zsave=[]
    for i,pb in enumerate(pbZ):
        zprod=pb*np.prod(pbR[in1[student_index],i])*np.prod(pbRzero[in0[student_index],i])
        #print(pb)
        if(i in z_index):
            zsave.append(zprod)
        sm+=zprod
    return sm,np.array(zsave),in1,in0#list(set(range(0,76))-set(in1)-set(in0))
```

```python
for i in range(0,258):
    sm,zsave,in1,in0=lhood_student_rating(pbZ,pbR,mvratings,pbRzero,i,[0])
    print(sm)
```

```python
def llhood(pbZ,pbR,mvratings,pbRzero,total,z_index):
    lsm=0
    for i in range(0,total):
        sm,zsave,in1,in0=lhood_student_rating(pbZ,pbR,mvratings,pbRzero,i,z_index)
        #print(sm)
        lsm+=np.log(sm)
    return lsm/total 
```

```python
ans=llhood(pbZ,pbR,mvratings,pbRzero,total,[0])
print(ans)
```

```python
def e_step(pbZ,pbR,mvratings,pbRzero,student_index,z_index):
    sm,zsave,in1,in0=lhood_student_rating(pbZ,pbR,mvratings,pbRzero,student_index,z_index)
    #print(zsave)
    return zsave/sm,in1,in0
```

```python
ans,in1,in0=e_step(pbZ,pbR,mvratings,pbRzero,4,[0,1,2,3])
#print(sm)
print(ans)
```

```python
def m_step(pbZ,pbR,mvratings,pbRzero,total):
    newpbZ=np.zeros(pbZ.shape).reshape((-1,1))
    #newpbR=np.zeros(pbR.shape)
    #smbz=np.zeros(pbZ.shape)
    #smbr=np.zeros(pbR.shape)
    one_sum=np.zeros(pbR.shape)
    zero_sum=np.zeros(pbR.shape)
    question_sum=np.zeros(pbR.shape)
    for i in range(0,total):
        pZr,in1,in0=e_step(pbZ,pbR,mvratings,pbRzero,i,range(0,4))
        #print(in1[0])
        newpbZ+=pZr.reshape((4,1))
        inquestion=list(set(range(0,76))-set(in1[i])-set(in0[i]))
        #print(inquestion)
        #print('indices ',inquestion)
        one_sum[in1[i],0:4]+=pZr.reshape((1,4))
        #print(in0[i])
        #zero_sum[in0[i],0:4]+=pZr.reshape((1,4))
        question_sum[inquestion,0:4]+=pZr.reshape((1,4))*pbR[inquestion,0:4]
    numerator=question_sum+one_sum
    #print('num ',numerator)
    denominator=newpbZ
    #print('den ',denominator)
    #print('new prob Z ',newpbZ)
    return newpbZ/total,numerator/denominator.reshape((1,4))
```

```python
pbZ=split_space(probZ)
pbZ=np.array(pbZ)
pbR=np.zeros((76,4))
for i,val in enumerate(probR):
    pbR[i,0:4]=val.split('   ')[1:5]
print(pbR)
ans1,ans2=m_step(pbZ,pbR,mvratings,1-pbR,total)
print('ans1: ',ans1)
print('ans2; ',ans2)
```

```python
lhood=[]
pbZ=split_space(probZ)
pbZ=np.array(pbZ)
probR=read_file('hw8_probR_init.txt')
pbR=np.zeros((76,4))
for i,val in enumerate(probR):
    pbR[i,0:4]=val.split('   ')[1:5]
lhood.append(llhood(pbZ,pbR,mvratings,pbRzero,total,[0]))
for i in range(0,256):
    pbZ,pbR=m_step(pbZ,pbR,mvratings,1-pbR,total)
    ans=llhood(pbZ,pbR,mvratings,1-pbR,total,[0,1,2,3])
    print(i)
    #print(ans)
    lhood.append(ans)
anslhood={i:x for i,x in enumerate(lhood)}
```

```python
indexlist=[0 if i==-1 else 2**i for i in range(-1,9)]
#indexlist=[0,1,2,4,5,]
tab=tt.Texttable()
tab.header(['iteration','log-likelihood'])
for x in indexlist:
    row=[x,str(anslhood[x])]
    tab.add_row(row)
tab.set_cols_align(['c','c'])
s=tab.draw()
print(s)
```
